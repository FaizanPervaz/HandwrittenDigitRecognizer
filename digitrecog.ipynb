```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
#build the model
model = tf.keras.models.Sequential() #basic sequential nueral network model
#flatten layer converts the 28x28 image into a 1x784 pixel input
model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
#dense layer is a fully connected layer like each nueorn is connected to each other
model.add(tf.keras.layers.Dense(units=128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(units=128, activation=tf.nn.relu))
#output layer has 10 neurons because we have 10 classes
model.add(tf.keras.layers.Dense(units=10, activation=tf.nn.softmax))
```

```python
# #compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# #train the model
model.fit(x_train, y_train, epochs=3)
```

```python
# #saving the model
model.save('HWDR.model')

```

```python
model = tf.keras.models.load_model('HWDR.model')
```

```python
# #testing the model
val_loss, val_acc = model.evaluate(x_test, y_test)
print(f"Loss : {val_loss}, Accuracy : {(val_acc*100)}%")
```

```python
image = 1
while os.path.isfile(f"Sample/img_{image}.jpg"):
    try:
        img = cv2.imread(f"Sample/img_{image}.jpg", cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (28, 28))
        img = img / 255.0  # Normalize the image
        img = np.reshape(img, (1, 28, 28))  # Reshape for model input
        prediction = model.predict(img)
        predicted_label = np.argmax(prediction)
        print(f"The {image} Number is probably: {predicted_label}")
        plt.imshow(img[0], cmap=plt.cm.binary)
        plt.show()
    except:
        print("Error!")
    finally:
        image += 1
```
